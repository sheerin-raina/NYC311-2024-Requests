```python
import pandas as pd

file_path = '/Users/sheerinraina/Downloads/311 Service Request 2024 Q1.csv'
df = pd.read_csv(file_path)


seen_rows = {}
rows_to_keep = []



for index, row in df.iterrows():
    key = (row['Incident Address'], row['Complaint Type'], row['Descriptor'], row['Incident Zip'])
    
    if key in seen_rows:
        original_index = seen_rows[key]
        original_desc = df.loc[original_index, 'Resolution Description']
        current_desc = row['Resolution Description']
        
        
        if 'duplicate' in str(current_desc).lower():
            # Current row has "duplicate" - keep original
            continue
        elif 'duplicate' in str(original_desc).lower():
            # Original has "duplicate" - replace with current
            rows_to_keep.remove(original_index)
            rows_to_keep.append(index)
            seen_rows[key] = index
        else:
            # Neither has "duplicate" - keep first one (original)
            continue
    else:
        # First time seeing this combination
        seen_rows[key] = index
        rows_to_keep.append(index)


df_clean = df.loc[rows_to_keep]

print(f"Before removing duplicates: {len(df)} rows")
print(f"After removing duplicates: {len(df_clean)} rows")
print(f"Removed: {len(df) - len(df_clean)} rows")

print(df_clean)
```
